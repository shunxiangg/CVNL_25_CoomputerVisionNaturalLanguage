```python
import numpy as np
x_np = np.random.random((4,4))
print(x_np)
```

```python
x_pt = torch.tensor(x_np)
print(x_pt)
print()
print(x_np.dtype, x_pt.dtype)
```

```python
x_np = np.asarray(x_np, dtype=np.float32)
x_pt = torch.tensor(x_np, dtype=torch.float32)
print(x_np.dtype, x_pt.dtype)
```

```python
b_np = (x_np > 0.5)
print(b_np)
print(b_np.dtype)
```

```python
b_pt = (x_pt > 0.5)
print(b_pt)
print(b_pt.dtype)
```
